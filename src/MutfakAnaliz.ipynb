#### Her Kuzinnin en iyi 4 malezemesi

```python
malzemesayisi=4
```

```python
##fonksiyon tanm 
def en_iyi_malzemeler(row):
    print(row.name.upper())
    row_sirala = row.sort_values(ascending=False)*100
    iyi_malzelemeler = list(row_sirala.index.values)[0:malzemesayisi]
    row_sirala = list(row_sirala)[0:malzemesayisi]

    for ind, ingredient in enumerate(iyi_malzelemeler):
        print("%s (%d%%)" % (ingredient, row_sirala[ind]), end=' ')
    print("\n")
```

```python
olusturKuzins = Kuzins.apply(en_iyi_malzemeler, axis=1)
```
